```python
import folium

import pandas as pd
```

```python
df = pd.read_excel("ind_cordinates.xlsx")
```

```python
df.columns
df.head(3)
```

```python
df['Cities'][2]
```

```python
cord = []
for x in df['Coordinates']:
    val = []
    val.append(float(x.split(',')[0]))
    val.append(float(x.split(',')[1]))
    cord.append(val)
    
cord
```

```python
map = folium.Map(location = [20.5937, 78.9629],
                zoom_start = 5)
map
```

```python
folium.Marker([20.5937, 78.9629],popup="<b>India</b>", tooltip="<i>India</i>").add_to(map)
map
```

```python
count = 0
for x in cord:
    folium.Marker(x, popup=str(x), tooltip="<b>"+df['Cities'][count]+"</b>").add_to(map)
    count += 1
 
```
