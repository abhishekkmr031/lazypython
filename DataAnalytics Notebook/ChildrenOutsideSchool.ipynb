```python
df = pd.read_csv("datafile.csv")
df.head(5)
```

```python
df_states = df.set_index("States")

df_states.fillna(0)
```

```python
########### Dictionary for indian states ##################

text = '''Bihar, India	25.0961	85.3131
Arunachal Pradesh, India	28.2180	94.7278
Madhya Pradesh, India	23.473324	77.947998
Haryana, India	29.238478	76.431885
Chhattisgarh, India	21.295132	81.828232
Haryana, India	29.065773	76.040497
Maharashtra, India	19.601194	75.552979
Tripura, India	23.745127	91.746826
Andhra Pradesh, India	17.874857	78.100815
Karnataka, India	15.317277	75.713890
Kerala, India	10.850516	76.271080
Uttar Pradesh, India	28.207609	79.826660
Assam, India	26.244156	92.537842
Maharashtra, India	19.663280	75.300293
Tamil Nadu, India	11.127123	78.656891
Karnataka, India	15.317277	75.713890
West Bengal, India	22.978624	87.747803
Gujarat, India	22.309425	72.136230
Orissa, India	20.940920	84.803467
Raja st ha n, India	27.391277	73.432617
Himachal Pradesh, India	32.084206	77.571167'''

csv = text.split('\n')
dic = dict()
states = []
for x in csv:
    dic[x.split("\t")[0].split(",")[0] ] =str(x.split("\t")[1]) +","+ str( x.split("\t")[2])
    states.append(x.split("\t")[0].split(",")[0])
```

```python
states
```

```python
dic
```

```python
df_states.fillna(0, inplace=True)
```

```python


for state in states:
    val = df_states.loc[state, "Hindus - Persons"]
    pop = "<b>No. of Children<br>Age 6 - 17<br>Outside of School  :" + str(val) + "</b>"
    clr = ""
    if val>=0 and val<=5:
        clr='blue'
    elif val>5 and val<=10:
        clr='green'
    elif val>10 and val<=15:
        clr='yellow'
    elif val>15 and val<=20:
        clr='orange'
    else:
        clr='red'
    
    folium.CircleMarker(radius = int(val)+5,
              location = [float(dic[state].split(',')[0]), float(dic[state].split(',')[1])],
              popup=pop,
              tooltip="<b>"+state+"</b>",
              color=clr,
              fill=True,
              fill_color=clr).add_to(map)
    
```
