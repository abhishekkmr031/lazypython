```python
df_can = pd.read_excel("Canada.xlsx", sheet_name ="Canada by Citizenship", skiprows=range(0, 20), skipfooter=2)
```

```python
df_can.head()
```

```python
years = list(range(1980, 2013))
```

```python
df_can.set_index('OdName', inplace=True)
df_can.index.values
```

```python
countries = df_can.index.values
print(countries)
```

```python
df_can.loc['China', years].plot(kind="line",color='b')

print(df_can.loc[['China', 'India'], years[1:5]])

```

```python
## plot line plot for 3 countries
plt.plot(df_can.loc['India', years], 'r', df_can.loc['China', years], 'b', df_can.loc['Haiti', years], 'b')
```

```python
## Remove the data which is not informative

df_can.drop(['Type', 'AREA', 'REG', 'DEV'], axis=1, inplace=True)
```

```python
print(df_can.head())
```

```python
df_can['Total'] = 0
```

```python
## print the name of all Columns
list(df_can)
```

```python
print(df_can.head())
```

```python
## creating a new column which have the Total of all the number of immigrants form 1980 - 2012
df_can['Total'] = 0

## A list which will contains the all values while calculating
total = []
for country in countries:
    total.append(df_can.loc[country, years].sum())

    
## Assigning total list to DataFrame[' Total ']
df_can['Total'] = total
print(df_can['Total'])
```

```python
print(df_can.head())
```

```python
df_can.sort_values(['Total'], ascending=False,  axis=0, inplace=True)
print(df_can.head())
```
