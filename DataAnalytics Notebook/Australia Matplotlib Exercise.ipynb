```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## importing data
df_can = pd.read_excel("Canada.xlsx",
                       sheet_name ="Canada by Citizenship",
                       skiprows=20,
                       skipfooter=1
                      )
```

```python
## Data Manipulation

## Drop unnecessary columns
df_can.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV'], axis=1, inplace=True)

## rename the Columns to make it readable
df_can.rename({'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region', 'DevName':'Development'}, axis=1, inplace=True)

## Fill Null values
df_can.fillna(value=0, axis=1)

##  Remove the Columns which have values = ..
##df_can.drop(list(map(int, range(2009, 2014))), axis=1, inplace=True)
```

```python
## Let's visualize the data
df_can.set_index('Country', inplace=True)
```

```python
years = list(map(int, range(1980, 2014)))
# Create a bar graph for country=India
df_can.loc['India', years].plot(kind='line')
print(df_can.loc[['India'], years[1:5]])

plt.show()
```

```python
#df_tot = pd.DataFrame(df_can[years].sum(axis=0))
years1 = list(map(int, range(1980, 2014)))
df_can.set_index( years1)
```

```python
# we can use the sum() method to get the total population per year
df_tot = pd.DataFrame(df_can[years1].sum(axis=0))

# change the years to type int (useful for regression later on)
df_tot.index = map(int, df_tot.index)

# reset the index to put in back in as a column in the df_tot dataframe
df_tot.reset_index(inplace = True)

# rename columns
df_tot.columns = ['year', 'total']

# view the final dataframe
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```
