```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_can = pd.read_excel("Canada.xlsx", 
                       sheet_name="Canada by Citizenship",
                      skiprows=20,
                      skipfooter=2)
```

```python
## Data Pre Processing

df_can.drop(['Type', 'Coverage', 'AREA', 'REG', 'DEV', 'DevName'], axis=1, inplace=True)
df_can.rename({'OdName':'Country', 'AreaName':'Continents', 'RegName':'Region'}, axis=1, inplace=True)
```

```python
df_can.dtypes
### years are in Int64

years = list(map(int, range(1980, 2014)))
```

```python
df_can.describe(include='all')
## Region = 22, Western Asia
## continents = 6, Africa
## country = 195
```

```python
df_can[df_can.Region=='Western Asia']
```

```python
df_can.set_index('Country', inplace=True)
```

```python
## Data Visualisation
### Bar Graph
df_can.loc['India', years].plot(kind='bar',
                               figsize=(20, 8))

plt.show()
```

```python
### Line Graph
df_can.loc['India', years].plot(kind='line',
                               figsize=(20, 8),
                               linewidth=3)
```

```python
df_can['Total'] = df_can.sum(axis=1)

#df_can.loc['India', 'Total']

#df_can.loc['India', years].mean()
```

```python
## Area Chart for top 5 Countries by total immigrant population from 1980 - 2013
df_can.sort_values('Total',ascending=False)

df_top = df_can.head(5)
df_top.set_index(years)
df_top = df_top[years].transpose()

df_top.plot(kind='area',
           alpha=0.7,
           stacked=False,
           figsize=(20, 8))
```

```python
## Pie Chart

df_pie = df_can.groupby('Continents',
                        axis=0,
                       ).sum()

df_pie['Total'].plot(kind='pie',
                    autopct='%.0f%%',
                    radius=1.75)
```

```python
## Scatter Plot for Country = India
#df_ind = pd.DataFrame(df_can['India'])

df_imm = pd.DataFrame(df_can[years].sum(axis=0))
```

```python
df_imm.index  = map(int, df_imm.index)

df_imm.reset_index(inplace = True)

df_imm.columns = ['year', 'total']
```

```python
df_imm.plot(kind='scatter',
           x='year',
           y='total',
           figsize=(20, 8))
```
