```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
labeled_images = pd.read_csv("train_image.csv")
```

```python
labeled_images
```

```python
### Images and label
images = labeled_images.iloc[0:5000, 1:]
labels = labeled_images.iloc[0:5000, :1]
```

```python
### Data Splitting into train and test data
### Training and Testing data

train_images, test_images, train_labels, test_labels = train_test_split(images, labels, train_size=0.8, random_state=0)
```

```python
i=1
img=train_images.iloc[i].as_matrix()
img=img.reshape((28,28))
plt.imshow(img,cmap='gray')
plt.title(train_labels.iloc[i,0])
```

```python
i = 1
img = train_images.iloc[1].as_matrix()
img = img.reshape((28, 28))
plt.imshow(img, cmap="gray")
plt.title(train_labels.iloc[i])
```

```python
plt.hist(train_images.iloc[1])
```

```python
classifier = svm.SVC()
```

```python
classifier.fit(train_images, train_labels.values.ravel())
```

```python
classifier.score(test_images, test_labels)
```
